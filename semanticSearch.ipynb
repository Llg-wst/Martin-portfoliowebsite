```python
import csv

with open ('/content/items.csv') as file:
  lines = csv.reader(file)

  documents = []
  metadatas = []
  ids = []
  id = 1

  for i, line in enumerate(lines):
    if i==0:
      continue

    documents.append(line[1])
    metadatas.append({"item_id": line[0]})
    ids.append(str(id))
    id+=1
```

```python
import chromadb
from chromadb.utils import embedding_functions
default_ef = embedding_functions.DefaultEmbeddingFunction()
chroma_client = chromadb.Client()
```

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-mpnet-base-v2")
collection = chroma_client.create_collection(name="my_collection", embedding_function = sentence_transformer_ef)
```

```python
collection.add(
    documents= documents,
    metadatas=metadatas,
    ids=ids
)
```

```python
collection = chroma_client.delete_collection(name="my_collection")
```

```python
results = collection.query(
    query_texts=["rice"],
    n_results=5,
    include= ['metadatas', 'embeddings', 'distances', 'documents']
)
```
